```python
######Working Fine
import cv2
import re
from keras.models import load_model
import numpy as np
from datetime import datetime

# Load pre-trained Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load pre-trained deep learning model for emotion recognition
emotion_model = load_model('.\\emotiondetector.h5')  # Provide the path to your pre-trained emotion model

# Define emotion labels
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']



def seconds_to_minutes_format(seconds):
    # Calculate minutes and seconds
    minutes, seconds = divmod(seconds, 60)
    
    # Calculate hours and remaining minutes
    hours, minutes = divmod(minutes, 60)

    # Format the result as HH:MM:SS
    result = "{:02}:{:02}:{:02}".format(int(hours), int(minutes), int(seconds))

    return result




def detect_emotion(face_roi):
    face_roi = cv2.cvtColor(face_roi, cv2.COLOR_BGR2GRAY)
    face_roi = cv2.resize(face_roi, (48, 48))
    face_roi = face_roi.astype('float') / 255.0
    face_roi = np.expand_dims(face_roi, axis=0)
    face_roi = np.expand_dims(face_roi, axis=-1)

    emotion_prediction = emotion_model.predict(face_roi)[0]
    emotion_index = np.argmax(emotion_prediction)
    emotion_label = emotion_labels[emotion_index]

    return emotion_label

def detect_confusion_in_live_video():
    cap = cv2.VideoCapture(0)  # Use 0 for default webcam, or provide the index of the desired webcam if you have multiple cameras

    # Record the start time
    start_time = datetime.now()

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Convert the frame to grayscale for face detection
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the frame
        faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        for (x, y, w, h) in faces:
            face_roi = frame[y:y+h, x:x+w]
            emotion = detect_emotion(face_roi)

            # Calculate elapsed time from start_time
            elapsed_time = datetime.now() - start_time
            timestamp = elapsed_time.total_seconds()
#             print(type( "{:.2f}".format(timestamp)))
            count_down=float((( "{:.2f}".format(timestamp)).split('.')[0]))
           
#             print()
            text = f"{seconds_to_minutes_format(count_down)}: {emotion}"

            cv2.putText(frame, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Display the live video frame
        cv2.imshow('Live Video', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Call the function to start live emotion detection
detect_confusion_in_live_video()
```
